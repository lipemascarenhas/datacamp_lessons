```python
# Build cars DataFrame
names = ['United States', 'Australia', 'Japan', 'India', 'Russia', 'Morocco', 'Egypt']
dr =  [True, False, False, False, True, True, True]
cpc = [809, 731, 588, 18, 200, 70, 45]
dict = { 'country':names, 'drives_right':dr, 'cars_per_cap':cpc }
cars = pd.DataFrame(dict)
print(cars)
```

```python
#selecionar uma coluna como DataFrame
print(cars[['country']])
```

```python
#selecionar múltiplas colunas como Dataframe
print(cars[['country', 'drives_right']])
```

```python
#selecionar observações a partir de seu index
print(cars[0:3])
```

```python
#loc[] is label-based, which means that you have to specify rows and columns based on their row and column labels.
# iloc[] is integer index based, so you have to specify rows and columns by their integer index
print('loc function single brackets')
print(cars.loc['AUS'])
print('')
print('iloc function single brackets')
print(cars.iloc[2])
```

```python
print('loc function double brackets')
print(cars.loc[['AUS']])
print('')
print('iloc function double brackets')
print(cars.iloc[[2]])
```

### Step 01

```python
# Import Numpy as np
import numpy as np

# Set seed
np.random.seed(123)

# Starting step
step = 50

# Roll the dice
dice = np.random.randint(1,7)
```

```python
# Finish the control construct
if dice <= 2 :
    step = step - 1
elif dice <=5 :
    step = step + 1
else :
    step = step + np.random.randint(1,7)
```

```python
# Print out dice and step
print("Dice: " + str(dice))
print("Step: " + str(step))
```

### Step 02

```python
# Import Numpy as np
import numpy as np

# Set seed
np.random.seed(123)
```

```python
# Initialize random_walk
random_walk = [0]

# Complete the ___
for x in range(100) :
    # Set step: last element in random_walk
    step = random_walk[-1]

    # Roll the dice
    dice = np.random.randint(1,7)

    # Determine next step
    if dice <= 2:
        step = step - 1
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    # append next_step to random_walk
    random_walk.append(step)
```

```python
# Print random_walk
print(random_walk)
```

### step 07 (clumsiness condition)

```python
# numpy and matplotlib imported, seed set

# Simulate random walk 250 times
all_walks = []
for i in range(250) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)

        # Implement clumsiness
        if np.random.rand() <=0.001 :
            step = 0

        random_walk.append(step)
    all_walks.append(random_walk)
```

```python
# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks))
plt.plot(np_aw_t)
plt.show()
```

```python
# numpy and matplotlib imported, seed set

# Simulate random walk 500 times
all_walks = []
for i in range(500) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        if np.random.rand() <= 0.001 :
            step = 0
        random_walk.append(step)
    all_walks.append(random_walk)
```
